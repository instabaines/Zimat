```python
#Importing necessary libraries
import pandas as pd
from pandas.api.types import is_numeric_dtype
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.preprocessing import OneHotEncoder,StandardScaler,MinMaxScaler
from sklearn.decomposition import PCA
from imblearn.under_sampling import TomekLinks
from imblearn.over_sampling import SMOTE
import imblearn
import xgboost as xgb
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV,KFold,StratifiedKFold,train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import f1_score
from sklearn.neural_network import MLPClassifier
from sklearn.base import BaseEstimator, TransformerMixin,ClassifierMixin
import re
```

```python
sample=pd.read_csv('./SampleSubmission.csv')
```

```python
missing=['','na','Nan','NAN','NaN']
train=pd.read_csv('./Train.csv')
test=pd.read_csv('./Test.csv',na_values=missing)
```

```python

test.join_date=np.where(test.join_date.isnull(),'1/5/2018',test.join_date)
train.join_date=np.where(train.join_date.isnull(),'1/5/2018',train.join_date)
```

```python
class DateTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass
    def fit(self, X, y=None):
        return self

    def transform(self,X):
        X['join_date'] = pd.to_datetime(X['join_date'], infer_datetime_format=True)
        X["Month"] = X['join_date'].dt.month
        X["Quarter"] = X['join_date'].dt.quarter
        X["Year"] = X['join_date'].dt.year
        X["Day"] = X['join_date'].dt.day
        X["Week"] = X['join_date'].dt.week
        X["Season"] = np.where(X["Month"].isin([3,4,5]),"Spring",
        np.where(X["Month"].isin([6,7,8]),"Summer",np.where(X["Month"].isin
            ([9,10,11]),"Fall",np.where(X["Month"].isin([12,1,2]),"Winter","None"))))
        X.drop(['join_date'],axis=1,inplace=True)
        return X
```

```python
date_trans=DateTransformer()
train=date_trans.fit_transform(train)
test=date_trans.fit_transform(test)
```

```python
train['sex'].replace({'F':0,'M':1},inplace=True)
test['sex'].replace({'F':0,'M':1},inplace=True)
```

```python
class BirthTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass
    def fit(self, X, y=None):
        return self

    def transform(self,X):
        X['birth_year'].astype(int)
        X["Agegroup"] = np.where(X["birth_year"].isin(range(1930,1960)),0,np.where(X["birth_year"].isin(range(1960,1991)),1,np.where(X["birth_year"].isin(range(1991,2022)),2,"None")))
        return X
bt=BirthTransformer()
```

```python
train=bt.fit_transform(train)
test=bt.fit_transform(test)
train.Agegroup=train.Agegroup.apply(int)
test.Agegroup=test.Agegroup.apply(int)
```

```python
train['marital_status']=np.where(train['marital_status']=='f','F',train['marital_status'])
list1=list(train.marital_status.unique())
dict_replace={i:list1.index(i) for i in list1}
train['marital_status'].replace(dict_replace,inplace=True)
test['marital_status'].replace(dict_replace,inplace=True)
```

```python
#test['marital_status'].replace(dict_replace,inplace=True)
train.drop(['ID'],axis=1,inplace=True)
```

```python
#Identifying the categorical and non categorical columns
cat=[]
non_cat=[]
for i in train:
    if train[i].dtype==object:
        cat.append(i)
    else:
        non_cat.append(i)



# creating instance of one-hot-encoder
enc = OneHotEncoder(handle_unknown='ignore')
enc_df = pd.DataFrame(enc.fit_transform(train[cat]).toarray())
# merge with main data on key values
newtrain=train[non_cat]
newtrain= newtrain.join(enc_df)
newtrain
```

```python
enc_df = pd.DataFrame(enc.transform(test[cat]).toarray())
# merge with main data on key values
newtest=test[non_cat]
newtest= newtest.join(enc_df)
newtest
```

```python
target=train.columns[6:-7]
```

```python
scaler=MinMaxScaler()
new_train_scaled=scaler.fit_transform(newtrain.drop(target,axis=1))
new_test_scaled=scaler.transform(newtest.drop(target,axis=1))
```

```python
df1={i:'x' for i in target}
df1=pd.DataFrame(df1,index=range(10000))
df1['user']=test.ID
df1
```

```python

user=list(test.ID)
for i in target:
    X=new_train_scaled
    y=newtrain[i]
    #Over sampling
    smote = SMOTE(sampling_strategy='minority')
    try:
        X_sm, y_sm = smote.fit_resample(X, y)
    except ValueError:
        smote = SMOTE(sampling_strategy='minority',k_neighbors=2) 
        X_sm, y_sm = smote.fit_resample(X, y)
    params = {"objective": "binary:logistic", # for classification
          "booster" : "gbtree",   # use tree based models 
          "eta": 0.01,   # learning rate
          "max_depth": 10,    # maximum depth of a tree
          "subsample": 1.0,    # Subsample ratio of the training instances
          "colsample_bytree": 0.7,   # Subsample ratio of columns when constructing each tree
          "silent": 1,   # silent mode
          "seed": 10 ,  # Random number seed
           "n_estimators":300,
        
          }
    num_boost_round = 3000

    dtrain = xgb.DMatrix(X_sm, y_sm)
    
    
    
    # train the xgboost model
    model = xgb.train(params, dtrain, num_boost_round)
    
    y_pred = model.predict(xgb.DMatrix(new_test_scaled))
    df1[i]=y_pred
```
